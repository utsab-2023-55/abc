```python
def basic_oper(no_of_head=None):
    if no_of_head==None:
        no_of_head=input('Number of heads=')
    else:
        pass
    try:
        no_of_head=int(no_of_head)
    except ValueError:
        print('Integer number between 1 to 1095')
        basic_oper()
    else:
        if 1<=no_of_head<=1095:
            no_of_tail=input('Number of tails')
            try:
                no_of_tail=int(no_of_tail)
            except ValueError:
                print('Integer number between 1 to 1095')
                basic_oper(no_of_head)
            else:
                if 1<=no_of_tail<=1095:
                    print('\033[1mHEAD OF THE DATASET:-\033[0m')
                    display(matches.head(no_of_head))
                    print('\033[1mTAIL OF THE DATASET:-\033[0m')
                    display(matches.tail(no_of_tail))
                    print('\033[1mALL COLUMNS NAME:-\033[0m')
                    display(matches.columns)
                    print('\033[1mSHAPE OF THE DATASET:-\033[0m')
                    display(matches.shape)
                    print('\033[1mDESCRIBE ABOUT THE DATASET:-\033[0m')
                    display(matches.describe())
                    print('\033[1mINFORMATION ABOUT THE DATASET:-\033[0m')
                    display(matches.info())
                else:
                    print('Integer number between 1 to 1095')
                    basic_oper(no_of_head)
        else:
            print('Integer number between 1 to 1095')
            basic_oper()
```

8.Year wise percentage of winning toss by a team 

```python
def year_wise_winning_toss(Year=None):
    if Year==None:
        season_list=set(matches_copy['season'])
        print('all',end=' or ')
        for year in set(matches_copy['season']):
            print(year,end='|')
        Year=input('Year=')
    else:
        season_list=set(matches_copy['season'])
    if Year.lower()=='all':
        Set=set(matches_copy['team1'])
        for st in Set:
            print(st,end=',')
        Team=input('Team=')
        if Team in Set:
            z=len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & (matches_copy['toss_winner']==Team)])/\
              len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team))])*100
            plt.pie(x=[z,100-z],labels=['Yes','No'],autopct='%1.1f%%')
            plt.title(label=f'Percentage of winning toss by {Team}')
        else:
            print('Choose Team from the above list')
            year_wise_winning_toss(Year)
    else:
        if Year in season_list:
            Set=set(matches_copy.loc[matches_copy['season']==Year]['team1'])
            for st in Set:
                print(st,end=',')
            Team=input('Team=')    
            if Team in Set:
                y=len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & (matches_copy['toss_winner']==Team) & \
                  (matches_copy['season']==Year)])/len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
                  (matches_copy['season']==Year)])*100
                plt.pie(x=[y,100-y],labels=['Yes','No'],autopct='%1.1f%%')
                plt.title(label=f'Percentage of winning toss by {Team} in {Year}')
            else:
                print('Choose Team from the above list')
                year_wise_winning_toss(Year)
        else:
            print('Choose Year from above season list')
            year_wise_winning_toss()
```

9.Year wise percentage of choosing bat and field 

```python
def year_wise_toss_decision(Year=None):
    if Year==None:
        season_list=set(matches_copy['season'])
        print('all',end=' or ')
        for year in set(matches_copy['season']):
            print(year,end='|')
        Year=input('Year=')
    else:
        season_list=set(matches_copy['season'])
    if Year.lower()=='all':
        Set=set(matches_copy['team1'])
        for st in Set:
            print(st,end=',')
        Team=input('Team=')
        if Team in Set:
            z=len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='bat')])/\
              len(matches_copy.loc[matches_copy['toss_winner']==Team])*100
            plt.pie(x=[z,100-z],labels=['Bat','Field'],autopct='%1.1f%%')
            plt.title(label=f'Percentage of choosing bat by {Team}')
        else:
            print('Choose Team from the above list')
            year_wise_toss_decision(Year)
    else:
        if Year in season_list:
            Set=set(matches_copy.loc[matches_copy['season']==Year]['team1'])
            for st in Set:
                print(st,end=',')
            Team=input('Team=')    
            if Team in Set:
                if len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['season']==Year)])==0:
                    print(f"In {Year} {Team} never won a toss in above dataset")
                else:
                    y=len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='bat') & \
                      (matches_copy['season']==Year)])/len(matches_copy.loc[(matches_copy['toss_winner']==Team) &\
                      (matches_copy['season']==Year)])*100
                    plt.pie(x=[y,100-y],labels=['Bat','Field'],autopct='%1.1f%%')
                    plt.title(label=f'Percentage of choosing bat by {Team} in {Year}')
            else:
                print('Choose Team from the above list')
                year_wise_toss_decision(Year)
        else:
            print('Choose Year from above list')
            year_wise_toss_decision()
```

10.Year wise percentage of winning after first bat or field

```python
def winning_after_bat_or_field(Year=None):
    if Year==None:
        season_list=set(matches_copy['season'])
        print('all',end=' or ')
        for year in set(matches_copy['season']):
            print(year,end='|')
        Year=input('Year=')
    else:
        season_list=set(matches_copy['season'])  
    if Year.lower()=='all':
        Set=set(matches_copy['team1'])
        for st in Set:
            print(st,end=',')
        Team=input('Team=')
        if Team in Set:
            a=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='runs')])/\
              len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team))]))*100
                    
            b=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
              (matches_copy['winner']!=Team) & (matches_copy['result']=='wickets')])/\
              len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team))]))*100
            
            c=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='bat') & \
                  (matches_copy['result']=='tie')]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='field') \
                  & (matches_copy['result']=='tie')]))/len(matches_copy.loc[(matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)]))*100
            
            d=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='wickets')])/\
               len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team))]))*100
            
            e=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
              (matches_copy['winner']!=Team) & (matches_copy['result']=='runs')])/\
              len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team))]))*100
            
            f=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='field') & \
                  (matches_copy['result']=='tie')]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='bat') \
                  & (matches_copy['result']=='tie')]))/len(matches_copy.loc[(matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)]))*100

            plt.pie(x=[a,b,c,d,e,f],labels=None,autopct='%1.1f%%')
            plt.title(label=f'Percentage of winning after first bat or field by {Team}')
            plt.legend(labels=['Win After First Bat','Loose After First Bat','Tie After First Bat',\
                                      'Win After First Field','Loose After First Field','Tie After First Field'],\
                       bbox_to_anchor=(1.5, 1),loc='upper right')
        else:
            winning_after_bat_or_field(Year)
    else:
        if Year in season_list:
            Set=set(matches_copy.loc[matches_copy['season']==Year]['team1'])
            for st in Set:
                print(st,end=',')
            Team=input('Team=')    
            if Team in Set:
                a=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='runs') & \
                  (matches_copy['season']==Year)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100
                    
                b=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
                  (matches_copy['winner']!=Team) & (matches_copy['result']=='wickets') & (matches_copy['season']==Year)])/\
                  len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100
                    
                c=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='bat') & \
                  (matches_copy['result']=='tie') & (matches_copy['season']==Year)]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='field') \
                  & (matches_copy['season']==Year) & (matches_copy['result']=='tie')]))/len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100
                    
                d=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='wickets') &\
                  (matches_copy['season']==Year)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100
            
                e=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
                  (matches_copy['winner']!=Team) & (matches_copy['result']=='runs') & \
                  (matches_copy['season']==Year)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100
                    
                f=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='field') & \
                  (matches_copy['season']==Year) & (matches_copy['result']=='tie')]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='bat') \
                  & (matches_copy['result']=='tie') & (matches_copy['season']==Year)]))/len(matches_copy.loc[((matches_copy['team1']==Team) | \
                  (matches_copy['team2']==Team)) & (matches_copy['season']==Year)]))*100

                plt.pie(x=[a,b,c,d,e,f],labels=None,autopct='%1.1f%%')
                plt.title(label=f'Percentage of winning after first bat or field by {Team} in {Year}')
                plt.legend(labels=['Win After First Bat','Loose After First Bat','Tie After First Bat',\
                                   'Win After First Field','Loose After First Field','Tie After First Field'],\
                           bbox_to_anchor=(1.5, 1),loc='upper right')
            else:
                print('Choose Team from the above list')
                winning_after_bat_or_field(Year)
        else:
            print('Choose Year from above list')
            winning_after_bat_or_field()
```

```python
def venuewise_winning_after_first_bat_or_field(Venue=None):
    if Venue==None:
        venue_set=set(matches_copy['venue'])
        for st in venue_set:
            print(st,end=' | ')
        Venue=input('Venue=')
    else:
        venue_set=set(matches_copy['venue'])
    if Venue in venue_set:
        team_set=set(matches_copy.loc[matches_copy['venue']==Venue]['team1']) | \
                 set(matches_copy.loc[matches_copy['venue']==Venue]['team2'])
        for st in team_set:
            print(st,end=',')
        Team=input('Team=')
        if Team in team_set:
            a=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='runs') & \
              (matches_copy['venue']==Venue)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
                    
            b=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
              (matches_copy['winner']!=Team) & (matches_copy['result']=='wickets') & (matches_copy['venue']==Venue)])/\
               len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
            
            c=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='bat') & \
              (matches_copy['result']=='tie') & (matches_copy['venue']==Venue)]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='field') \
              & (matches_copy['venue']==Venue) & (matches_copy['result']=='tie')]))/len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
                    
            d=(len(matches_copy.loc[(matches_copy['winner']==Team) & (matches_copy['result']=='wickets') & \
              (matches_copy['venue']==Venue)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
            
            e=(len(matches_copy.loc[((matches_copy['team1']==Team) | (matches_copy['team2']==Team)) & \
              (matches_copy['winner']!=Team) & (matches_copy['result']=='runs') & \
              (matches_copy['venue']==Venue)])/len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
                    
            f=((len(matches_copy.loc[(matches_copy['toss_winner']==Team) & (matches_copy['toss_decision']=='field') & \
              (matches_copy['venue']==Venue) & (matches_copy['result']=='tie')]) + len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['toss_winner']!=Team) & (matches_copy['toss_decision']=='bat') \
              & (matches_copy['result']=='tie') & (matches_copy['venue']==Venue)]))/len(matches_copy.loc[((matches_copy['team1']==Team) | \
              (matches_copy['team2']==Team)) & (matches_copy['venue']==Venue)]))*100
            
            plt.pie(x=[a,b,c,d,e,f],labels=None,autopct='%1.1f%%')
            plt.title(label=f'Percentage of winning after first bat or field by {Team} in {Venue}')
            plt.legend(labels=['Win After First Bat','Loose After First Bat','Tie After First Bat',\
                              'Win After First Field','Loose After First Field','Tie After First Field'],\
                               bbox_to_anchor=(1.5, 1),loc='upper right')
        else:
            print('Select Team from above list')
            venuewise_winning_after_first_bat_or_field(Venue)
    else:
        print('Select Venue from above list')
        venuewise_winning_after_first_bat_or_field()
```
